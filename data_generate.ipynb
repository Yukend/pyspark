```python
###### PATIENT DATA GENERATOR ######


from concurrent.futures import ThreadPoolExecutor
from faker import Faker
import random

fake = Faker()
file = open("patient.csv", "a")


def generate_patient_data(_):
    patient_id = "PAT" + str(random.randrange(10000, 99999))
    first_name = fake.first_name()
    last_name = fake.last_name()
    date_of_birth = fake.date_of_birth(minimum_age=18, maximum_age=80).strftime('%Y-%m-%d')
    gender = random.choice(['Male', 'Female'])
    phone_number = random.randint(6000000000, 9999999999)
    email = fake.email()
    diagnosis = random.choice(['Headache', 'Fever', 'Cough', 'Back Pain', 'Sprained Ankle', 'Sore Throat', 'Cold', 'Fatigue', 'Hypertension'])
    admission_date = fake.date_this_decade().strftime('%Y-%m-%d')
    discharged = random.choice([True, False])
    organization = "ORG" + str(random.randint(1000, 1999))
    practitioner = "PRAC" + str(random.randint(100, 199))
    address = "ADRS" + str(random.randint(10000, 24999))


    return f"{patient_id},{first_name},{last_name},{date_of_birth},{gender},{address},{phone_number},{email},{diagnosis},{admission_date},{discharged},{organization},{practitioner}"

if __name__ == "__main__":
    num_records = 10000

    with ThreadPoolExecutor() as executor:
        # Use list comprehension to generate 10,000 records using 10 threads
        insert_queries = list(executor.map(generate_patient_data, range(num_records)))

    # Print the generated insert queries
    for query in insert_queries:
        file.write(query + "\n")
```

```python
### Generate synthetic medicine data and write to CSV file

from concurrent.futures import ThreadPoolExecutor
from faker import Faker
import random
from datetime import datetime, timedelta

# Create a Faker instance
fake = Faker()
file = open("medicine.csv", "a")

# Function to generate synthetic medicine data
def generate_medicine_data(_):
    medicine_id = "MED" + str(random.randrange(100, 299))
    medicine_name = random.choice(medicine_list)
    manufacturer = random.choice(manufacturers)
    dosage_form = random.choice(["Tablet", "Tonic", "Injection", "Powder", "Liquids"])
    strength = round(random.uniform(1, 100), 2)
    unit = random.choice(["100mg", "200mg", "300mg", "400mg", "500mg"])
    quantity = random.randint(1, 1000)
    price = round(random.uniform(1, 1000), 2)
    expiry_date = (datetime.now() + timedelta(days=random.randint(30, 365))).strftime('%Y-%m-%d')

    return f"{medicine_id},{medicine_name},{manufacturer},{dosage_form},{strength},{unit},{quantity},{price},{expiry_date}"

if __name__ == "__main__":
    num_medicines = 200

    with ThreadPoolExecutor() as executor:
        # Use list comprehension to generate 20,000 diagnosis records using 10 threads
        diagnosis_records = list(executor.map(generate_medicine_data, range(num_medicines)))

    # Print the generated diagnosis records
    for record in diagnosis_records:
        file.write(record + "\n")

    file.close()
```

```python
import csv
import random

# Create a mapping object for patients, medicines, and diagnoses
def generate_patient_medicine_diagnosis_mapping(num_mappings):
    patient_medicine_diagnosis_mapping = []
    for _ in range(num_mappings):
        patient_id = "PAT" + str(random.randint(10000, 19999))
        medicine_id = "MED" + str(random.randint(100, 299))
        diagnosis_id = "DIAG" + str(random.randint(10000, 39999))
        quantity = random.randint(1, 10)
        dosage = random.choice(["Once a day", "Twice a day", "Thrice a day"])
        patient_medicine_diagnosis_mapping.append((patient_id, medicine_id, diagnosis_id, quantity, dosage))
    return patient_medicine_diagnosis_mapping

# Generate 10,000 patient-medicine-diagnosis mappings
patient_medicine_diagnosis_mapping = generate_patient_medicine_diagnosis_mapping(10000)

# Write the mapping to a CSV file
with open("medorder.csv", "w", newline="") as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(["Patient ID", "Medicine ID", "Diagnosis ID", "Quantity", "Dosage"])
    writer.writerows(patient_medicine_diagnosis_mapping)

print("Patient-medicine-diagnosis mapping written to patient_medicine_diagnosis_mapping.csv")
```
